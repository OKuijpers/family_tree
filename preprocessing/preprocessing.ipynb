```python
def create_new_person(name, partner_name, partner_id, df):
    # Generate a new ID
    new_id = df['ID'].max() + 1 if not df['ID'].empty else 1
    
    # Create a new row for the new person
    new_person = pd.DataFrame({
        'Name': [name],
        'Partner': [partner_name],  # No partner info for the new person
        'ID': [new_id],
        'Partner_id': [partner_id]  # No partner ID initially
    })
    
    # Append the new person to the DataFrame
    df = pd.concat([df, new_person], ignore_index=True)
    
    return df

for idx, row in df.iterrows():
    if pd.isnull(row['Partner_id']) and not pd.isnull(row['Partner']):
        partner_row = df[df['Name'] == row['Partner']]
        
        if partner_row.empty:
            df = create_new_person(row['Partner'], row['Name'], row['ID'], df)
            partner_id = df['ID'].max()
            df.loc[idx, 'Partner_id'] = partner_id
            continue
            
        if len(partner_row) != 1:
            partner_row = partner_row[partner_row['Partner'] == row['Name']]
            assert len(partner_row) == 1
        df.loc[idx, 'Partner_id'] = partner_row['ID'].values[0]
df        
```

```python
def create_new_person(name, partner_name=None, partner_id=None, father_name=None, mother_name=None, father_id=None, mother_id=None, df=None, sex=None):
    if df is None:
        raise ValueError("DataFrame must be provided")

    # Generate a new ID
    new_id = df['ID'].max() + 1 if not df['ID'].empty else 1
    
    # Create a new row for the new person
    new_person = pd.DataFrame({
        'Name': [name],
        'Partner': [partner_name],
        'ID': [new_id],
        'Partner_id': [partner_id],
        'Sex': [sex],
        'Mother': [mother_name],
        'Father_id': [father_id],
        'Mother_id': [mother_id]
    })
    # Append the new person to the DataFrame
    df = pd.concat([df, new_person], ignore_index=True)
    
    return df


def handle_partner_id(row, df):
    if pd.isnull(row['Partner_id']) and not pd.isnull(row['Partner']):
        partner_row = df[df['Name'] == row['Partner']]

        if partner_row.empty:
            df = create_new_person(
                row['Partner'],
                partner_name=row['Name'],
                partner_id=row['ID'],
                df=df
            )
            partner_id = df['ID'].max()
            df.loc[row.name, 'Partner_id'] = partner_id
            return df

        if len(partner_row) != 1:
            partner_row = partner_row[partner_row['Partner'] == row['Name']]
            assert len(partner_row) == 1
        df.loc[row.name, 'Partner_id'] = partner_row['ID'].values[0]
    return df

def handle_mother_id(row, df):
    if pd.isnull(row['Mother_id']) and not pd.isnull(row['Mother']):
        mother_row = df[df['Name'] == row['Mother']]

        if mother_row.empty:
            df = create_new_person(
                row['Mother'],
                sex='Vrouw',
                partner_id=row['Father_id'],
                df=df
            )
            mother_id = df['ID'].max()
            df.loc[row.name, 'Mother_id'] = mother_id
            return df

        if len(mother_row) != 1:
            mother_row = mother_row[mother_row['Mother'] == row['Name']]
            assert len(mother_row) == 1
        df.loc[row.name, 'Mother_id'] = mother_row['ID'].values[0]
    return df

def handle_father_id(row, df):
    if pd.isnull(row['Father_id']) and not pd.isnull(row['Father']):
        father_row = df[df['Name'] == row['Father']]

        if father_row.empty:
            df = create_new_person(
                row['Father'],
                sex='Male',
                partner_id=row['Mother_id'],
                df=df
            )
            father_id = df['ID'].max()
            df.loc[row.name, 'Father_id'] = father_id
            return df

        if len(father_row) != 1:
            father_row = father_row[father_row['Father'] == row['Name']]
            assert len(father_row) == 1
        df.loc[row.name, 'Father_id'] = father_row['ID'].values[0]
    return df

# Main loop
for idx, row in df.iterrows():
    df = handle_partner_id(row, df)
    df = handle_mother_id(row, df)
    df = handle_father_id(row, df)
df
```

```python
df['Sex'] = df['Sex'].replace({'Man':'Male', 'Vrouw': 'Female'})
```

```python
from datetime import datetime
def convert_to_ddmmyyyy(date_str):
    if pd.isna(date_str):
        return None
    
    # Try parsing with different formats
    for fmt in ('%d-%m-%Y', '%d-%m-%y', '%Y-%m-%d', '%m-%d-%Y'):
        try:
            return datetime.strptime(date_str, fmt).strftime('%d-%m-%Y')
        except ValueError:
            continue
    
    # If no format matches, return the original value (or handle error)
    return date_str



final_int_cols = ['Father_id', 'Mother_id', 'Partner_id']
df[final_int_cols] =df[final_int_cols].fillna(-1).astype(int)
final_date_cols = ['DateOfBirth', 'DateOfDeath', 'DateOfMarriage', 'DateOfBaptism']
URL_cols = ['URLFather','URLMother','URLPartner','URLProfession', 'URLDateOfBirth', 'URLPlaceOfBirth', 'URLDateOfDeath', 'URLPlaceOfDeath']
# Apply the conversion function to each date column
for col in final_date_cols:
    df[col] = df[col].apply(convert_to_ddmmyyyy)
df
```

```python
final_cols = ['ID', 'Name', 'Sex'] + final_int_cols + final_date_cols + ['PlaceOfBirth', 'PlaceOfDeath', 'PlaceOfMarriage', 'PlaceOfBaptism']
final_cols += ['Profession'] + URL_cols
df = df[final_cols]
```

```python
def extract_last_name(full_name):
    parts = full_name.split()
    tussenvoegsels = {'van', 'der', 'de', 'den', 'het', 'ten', 'ter'}

    # Initialize the last name and tussenvoegsel
    last_name_parts = []
    for part in reversed(parts):
        if part.lower() in tussenvoegsels and last_name_parts:
            last_name_parts.insert(0, part)
        elif not last_name_parts:
            last_name_parts.insert(0, part)
        else:
            break

    last_name = last_name_parts.pop()
    tussenvoegsel = ' '.join(last_name_parts)

    if tussenvoegsel:
        return f"{last_name}, {tussenvoegsel}"
    else:
        return last_name

df['Surname'] = df['Name'].apply(extract_last_name)
df
```
